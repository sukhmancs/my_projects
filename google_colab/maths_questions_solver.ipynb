```python
#@title Find missing digit (or digits)
def number_guesser():
    original_number = input("please enter the divident here: ")
    number = original_number
    divisor = int(input("please input the divisor: "))
    count = 0
    missing_digits = 0
    num_list = set([])
    for i, s in enumerate(number):
      if s.isalpha() == True and count == 0:
        number = number.replace(number[i], "x")
        count += 1
        missing_digits += 1
      elif s.isalpha() == True and count > 0:
        number = number.replace(number[i], "t")
        missing_digits += 1
    if missing_digits > 2:
      print(f"\033[93;40;4mWARNING: 2 missing digits expected {missing_digits} provided."
              " The number of possible outcomes could be lesser than expected\n\033[0m")
    for i in range(10):
      for j in range(10):
        guess = number.replace("x", str(i))
        guess = guess.replace("t", str(j))
        num = int(guess)
        if num % divisor == 0:
            #print(f"The missing digit is {i},{j} and the numbers are {num}.")
            num_list.add(num)
    print('\033[94m' + '\033[4m' + "Numbers list is ", num_list, '\033[0m')
    print(f"No other missing digit was found that makes the {original_number} divisible by {divisor}.")

number_guesser()
```

```python
#@title Find permutations
from itertools import permutations

def find_permutations(iterable):
    return permutations(iterable)

iterable = input("Please enter list of numbers seperated by spaces: ").split()
result = find_permutations(iterable)

print("Number of permutations: ", len(list(result)))

## Uncomment to see the actual permutation values
#for permutation in result:
#   print(permutation)
```

```python
#@title Find number of combinations
import itertools

def find_combinations(items, n):
    return list(itertools.combinations(items, n))

items = input("Please enter list of values separated with spaces (n): ")
n = int(input("Size of combinations (r): "))

combinations = find_combinations(items, n)
print("Number of combinations: ", len(list(combinations)))
```

```python
#@title Find Congruence

def is_congruence(dividend, divisor, remainder):
  results = (dividend - remainder) / divisor
  if (abs(results - int(results)) == 0):
    results = int(results)
  if (type(results) is int):
    return True
  return False

dividend = int(input("Enter the dividend: "))
divisor = int(input("Enter the divisor: "))
remainder = dividend % divisor

print("\nRemainder:", remainder)
print("Is Congruent:", is_congruence(dividend, divisor, remainder))
```
