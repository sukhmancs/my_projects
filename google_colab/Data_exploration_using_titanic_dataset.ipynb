# __Delete variables__:
Delete only those variables where NAN values are >= 50% of total number of values. 
For example, in our case, it is column 'Cabin' which contains more than 50% of the NAN values. 

```python
#delete 'Cabin' column
#treat 'age', 'Embarked' normally 
```

```python
train_full_d = train_full.drop('Cabin', axis=1)
train_full_d.head()
```

#__Impute Values__:
Variables that contain few NAN values must be treated with imputation (filling NAN traps). We can use `SimpleImputer` model to first train (fit) on our data and then transform our values. 

```python
# for now remove categorical columns (simplicity)
train_full_dEo = train_full_d.select_dtypes(exclude=['object'])
train_full_dEo.head()
```

```python
# impute values (in this case only numeric values)
from sklearn.impute import SimpleImputer

# Imputation
my_imputer = SimpleImputer() # calling 
# fit on our model and transform
imputed_X_train_full = pd.DataFrame(my_imputer.fit_transform(train_full_dEo))

imputed_X_train_full.head()
imputed_X_train_full.columns = train_full_dEo.columns
# see if there any NAN trap left in 'age' column.
#imputed_X_train_full['Age'].isnull().any()
imputed_X_train_full 
```

```python
# Importing
import sklearn
#from sklearn.datasets import load_boston
import pandas as pd
import numpy as np

# Load the dataset
#bos_hou = load_boston()

# Create the dataframe
#column_name = bos_hou.feature_names
#df_boston = pd.DataFrame(bos_hou.data)
#df_boston.columns = column_name
#df_boston.head()

''' Detection '''
# IQR 
# calculate 25th and 75th percentile --- means calculate 
# those numbers below which 25% and 75% values can be found.
Q1 = np.percentile(imputed_X_train_full['Fare'], 25,
				interpolation = 'midpoint')

Q3 = np.percentile(imputed_X_train_full['Fare'], 75,
				interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", imputed_X_train_full.shape)

# Upper bound
upper = np.where(imputed_X_train_full['Fare'] >= (Q3+1.5*IQR))
print('upper[0]',upper[0])
# Lower bound
lower = np.where(imputed_X_train_full['Fare'] <= (Q1-1.5*IQR))
print(upper)
''' Removing the Outliers '''
imputed_X_train_full.drop(upper[0], inplace = True) # upper[0] because .drop function only accept list-like inputs
imputed_X_train_full.drop(lower[0], inplace = True) # inplace means not to produce another copy of dataframe but drop
																										# rows or columns from the original dataframe. 

print("New Shape: ", imputed_X_train_full.shape)
```

```python
imputed_y_train_full = imputed_X_train_full['Survived']
imputed_X_train_full_c = imputed_X_train_full.drop('Survived', axis=1)
```

```python
from sklearn.model_selection import train_test_split
# Divide data into training and validation subsets
X_train_full, X_valid_full, y_train, y_valid = train_test_split(imputed_X_train_full_c,
                                                                imputed_y_train_full, 
                                                                train_size=0.8,
                                                                test_size=0.2, random_state=0)
```

```python
# number of rows we have
len(X_train_full)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, 
                              random_state=0)
# fit model on training data
model.fit(X_train_full,y_train)
```

```python
import xgboost as xgb
xgb_model = xgb.XGBRegressor(n_estimators=500,
                             learning_rate=0.01,
                             )
xgb_model.fit(X_train_full,y_train)
```

```python
# predict on validation data
predics = xgb_model.predict(X_valid_full)
predics
```

```python
# mean_absolute_error
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_valid,predics)
```
